<h2 style="font-size: 20px; color: LightBlue;">The data does not always come out of memory in the format we want it in right out of the bat. Sometimes we have to do some more work ourselves to reformat it for the task at hand.</h2>

<h2 style="font-size: 25px; color: yellow;">map()</h2>

```python
points_mean_value = reviews['points'].mean()
```

```python
reviews['points'] = reviews['points'].map(lambda p : p - points_mean_value)
```

```python
reviews['points']
```

<h2 style="font-size: 25px; color: yellow;">apply()</h2>

```python
def modify_points_values(row):
    row['points'] = row['points'] + points_mean_value
    return row

reviews = reviews.apply(modify_points_values, axis = 'columns')
```

```python
reviews['points']
```

```python
reviews['points'] = reviews['points'] - points_mean_value
```

```python
reviews['points']
```

<h1 style = color:#33FF00>Grouping and Sorting</h1>

```python
reviews.head(2)
```

```python
# THIS GROUPS THE DATA ABOUT THE WINES WITH RESPECT TO THE POINTS
# THEN FOR EACH GROUP WE DETERMINED THE TOTAL POINTS 
reviews.groupby('points').points.count()
```

```python
# THIS GROUPS THE DATA ABOUT THE WINES WITH RESPECT TO THE POINTS
# THEN FOR EACH GROUP WE DETERMINED THE WINE WITH LOWEST PRICE
reviews.groupby('points').price.min()
```

```python
# HERE DATA IS GROUPED BY THE COUNTRIES
# LENGTH, MINIMUM PRICE, AND MAXIMUM PRICE IS ALSO DETERMINED FOR EACH GROUP
reviews_by_country = reviews.groupby(['country']).price.agg([len, "min", "max"])
reviews_by_country.sample(5)
```

<h2 style="font-size: 25px; color: yellow;">Sorting</h2>

```python
reviews_by_country = reviews_by_country.reset_index()
```

```python
reviews_by_country.sort_values(by="min").head(10)
```

```python
reviews_by_country.sort_values(by=['min', 'max']).head(10)
```

```python
reviews_by_country.sort_index(ascending=False).head(10)
```

<h2 style="font-size: 25px; color: yellow;">Combining</h2>

<h2 style="font-size: 20px; color:LightBlue;">The simplest combining method is concat(). Given a list of elements, this function will smush those elements together along an axis.</h2>

```python
videos_canada = pd.read_csv(r'D:\Python\Pandas\CAvideos.csv')
videos_britain = pd.read_csv(r'D:\Python\Pandas\GBvideos.csv')
```

```python
pd.concat([videos_britain, videos_canada]).sample(3)
```

<h2 style="font-size: 20px; color:LightBlue;">join() lets you combine different DataFrame objects which have an index in common. </h2>

```python
left = videos_canada.set_index(['title', 'trending_date'])
right = videos_britain.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_GB').head(2)
```
