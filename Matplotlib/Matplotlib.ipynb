```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

<h2 style= "font_size:25px; color:Yellow;">Scatter Plots</h2>

```python
x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
```

```python
plt.scatter(x_data, y_data)
plt.show()
```

#### Changing color of the plot

```python
plt.scatter(x_data, y_data, c="Red")
```

#### Changing the dots to *

```python
plt.scatter(x_data, y_data, marker='*')
```

#### Changing the size of the dots

```python
plt.scatter(x_data, y_data, s=100)
```

#### Adding transparency in the data plots

```python
data_x = np.random.random(1500) * 100
data_y = np.random.random(1500) * 100
```

```python
plt.scatter(data_x, data_y)
```

""" Here the graph has a lot of overlapping points, transparency may help visualize the graph a bit better """

```python
plt.scatter(data_x, data_y, alpha=0.3)
```

<h2 style= "font_size:25px; color:Yellow;">Bar Plots</h2>

```python
x = ["C++", "C", "Python", "Java", "Go"]
y = [50, 20, 100, 10, 35]
```

```python
plt.bar(x, y)
```

#### Changing color

```python
plt.bar(x, y, color="Grey")
```

```python
colors = ["y", "b", "m", "g"]
plt.bar(x, y, width=0.5, color = colors)
plt.show()
```

#### Changing the alignment of the bars w.r.t x-axis

```python
plt.bar(x, y, align="edge")
```

#### Changing the width of the bars

```python
plt.bar(x, y, width=0.3)
```

#### Changing the edge color of the bars

```python
plt.bar(x, y, edgecolor="black")
```

#### Making the edges more visible

```python
plt.bar(x, y, edgecolor="black", lw=5)
```

#### Changing the font size

```python
plt.bar(x, y)
plt.xlabel("Language", fontsize = 20)
plt.ylabel("Votes", fontsize = 20)
plt.title("Language Popularity Graph", fontsize = 25)
plt.tight_layout()
plt.show()
```

<h2 style= "font_size:25px; color:Yellow;">Histograms</h2>

```python
# generating ages data using normal distribution
ages = np.random.normal(20, 1.5, 1000)
# the data has 1000 values, mean = 20 and standard deviation = 1.5
```

```python
plt.hist(ages)
plt.show()
```

#### Changing the number of bins

```python
plt.hist(ages, bins=(ages.min(), 18, 21, ages.max()))
plt.show()
```

```python
plt.hist(ages, bins=10)
plt.show()
```

#### Plotting cumulative histogram

```python
plt.hist(ages, cumulative=True)
plt.show()
```

<h2 style= "font_size:25px; color:Yellow;">Pie Charts</h2>

```python
langs = ["Python", "C++", "Java", "C#", "Go"]
votes = [40, 20, 30, 5, 12]
```

```python
plt.pie(votes, labels=langs)
plt.show()
```

#### Exploding a section from the chart


```python
explodes = [0, 0, 0, 0.2, 0]
plt.pie(votes, labels=langs, explode=explodes)
plt.show()
```

#### Writing percentage in the charts

```python
plt.pie(votes, labels=langs, autopct="%.1f%%")
plt.show()
```

#### Changing the distance of the percentage

```python
plt.pie(votes, labels=langs, autopct="%.2f%%" ,pctdistance=0.8)
plt.show()
```

#### Changing the start angle of the chart

```python
plt.pie(votes, labels=langs, startangle=90)
plt.show()
```

<h2 style= "font_size:25px; color:Yellow;">Plot Customization</h2>

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
income = [55, 56, 62, 61, 72, 72, 73, 75]
income_ticks = list(range(50, 81, 2))
```

```python
plt.plot(years, income)
plt.show()
```

#### Setting the title of the plot

```python
plt.plot(years, income)
plt.title("Income of John ($)")
plt.show()
```

#### Setting x and y labels

```python
plt.plot(years, income)
plt.title("Income of John")
plt.xlabel("Year")
plt.ylabel("Income in $")
plt.show()
```

#### Adding y-ticks

```python
plt.plot(years, income)
plt.title("Income of John")
plt.xlabel("Year")
plt.ylabel("Income in $")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.show()
```

#### Changing the font sizes

```python
plt.plot(years, income)
plt.title("Income of John", fontsize = 20)
plt.xlabel("Year", fontsize = 13)
plt.ylabel("Income in $", fontsize = 13)
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.show()
```

#### Changing the font

```python
plt.plot(years, income)
plt.title("Income of John", fontsize = 20, fontfamily = "Bell MT")
plt.xlabel("Year", fontsize = 13, fontfamily = "Cascadia Mono")
plt.ylabel("Income in $", fontsize = 13, fontfamily = "Cascadia Mono")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks], fontfamily = "Cascadia Mono")
plt.xticks(fontfamily = "Cascadia Mono")
plt.show()
```

<h2 style= "font_size:25px; color:Yellow;">Legends and Multiple Plots</h2>

```python
stock_a = [100, 102, 91, 99, 101, 100, 105]
stock_b = [89, 93, 95, 100, 109, 106, 100]
stock_c = [115, 110, 108, 113, 100, 90, 95]
```

#### Adding a legend to the plot

A legend allows us to differentiate between multiple plots in the same graph

```python
plt.plot(stock_a, label = "comapny 1")
plt.plot(stock_b, label = "comapny 2")
plt.plot(stock_c, label = "comapny 3")
plt.legend()
plt.show()
```

#### Changing the location of the legend in the plot

```python
plt.plot(stock_a, label = "comapny 1")
plt.plot(stock_b, label = "comapny 2")
plt.plot(stock_c, label = "comapny 3")
plt.legend(loc = "lower right")
plt.show()
```

<h2 style= "font_size:25px; color:Yellow;">Sub Plots</h2>

#### Multiple plots in the same window

```python
x = np.arange(100)
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random Function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Log Function")

fig.suptitle("Four Plots")

plt.show()
```

```python
x = np.arange(100)
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random Function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Log Function")

fig.suptitle("Four Plots")

plt.tight_layout()

plt.savefig("Four Plots.png", dpi = 300, bbox_inches = "tight")
```

<h2 style= "font_size:25px; color:Yellow;">3D Plots</h2>

```python
ax = plt.axes(projection = "3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title("3D Plot")

plt.show()
```

#### Labeling the axis

```python
ax = plt.axes(projection = "3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title("3D Plot")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.tight_layout()

plt.show()
```

```python
ax = plt.axes(projection = "3d")

x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)
ax.set_title("3D Plot")

plt.show()
```

#### Surface Plots

```python
ax = plt.axes(projection = "3d")

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(Y) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap = "Spectral")
ax.set_title("3D Plot")

plt.show()
```
